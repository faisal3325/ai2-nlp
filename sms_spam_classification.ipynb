```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv("SMSSpamCollection.tsv", sep="\t")
```

```python
dataframe.head()
```

```python
message_length = []

for index, row in dataframe.iterrows():
    length_message_text = len (row.message)
    message_length.append(length_message_text)
```

```python
message_length
```

```python
dataframe['length'] = message_length
```
